```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Market_Basket_Optimisation.csv', header = None) #No headers/Column names
#Recreate the dataset to create the dataframe/format that apriori expects
#Initialize a list and populate it by looping all the data
transactions = []
for i in range(0,7501):
  transactions.append([str(dataset.values[i,j])for j in range(0,20)])
  #make sure everything in list are strings
```

### **Training the Apriori model on the dataset**

```python
#Calling apriori function from apyori
from apyori import apriori #rules are output of this function
rules = apriori(transactions = transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length =2 ) #it does some association/correlation rules
```

lift less than 3 does not give relevant rules
If buy 2 get 2 free min-length and max-length to 2 (according to business)

### **Putting the results well organised into a pandas dataframe**

```python
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```

reuslts -- > rules;
left hand side-- full list of results
and access them one by one using indices
and for right hand side -- full list of results and access them one by one using indices;
New dataframe which takes input of everything
